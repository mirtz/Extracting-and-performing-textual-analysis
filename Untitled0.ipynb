```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import requests
import nltk
from nltk import sent_tokenize,word_tokenize
from urllib.parse import urlparse
from nltk.corpus import words,stopwords
import string
from nltk.sentiment.vader import SentimentIntensityAnalyzer


nltk.download('punkt')
nltk.download('words')
nltk.download('cmudict')
nltk.download('stopwords')
nltk.download('vader_lexicon')


def avg_words_per_sentence(url):
  response = requests.get(url)
  if response.status_code == 200:
    webpage_text = response.text
    # Tokenize the text into sentences
    sentences = sent_tokenize(webpage_text)
    # Calculate the average number of words per sentence
    total_words = 0
    total_sentences = len(sentences)

    for sentence in sentences:
        words = word_tokenize(sentence)
        sentence_length = len(words)
        total_words += sentence_length

    average_words_per_sentence = total_words / total_sentences
    return average_words_per_sentence

def count_complex_words(url):
    response = requests.get(url)
    webpage_text = response.text
    words = word_tokenize(webpage_text)# Tokenize the text into words
    # Create a pronunciation dictionary using the CMU Pronouncing Dictionary
    pronunciation_dict = nltk.corpus.cmudict.dict()
    # Initialize a count for complex words
    complex_word_count = 0
    for word in words:
        # Remove punctuation and convert to lowercase for consistent matching
        word = word.lower()
        word = ''.join(e for e in word if e.isalnum())
        # Use the pronunciation dictionary to count syllables
        syllables = pronunciation_dict.get(word)
        if syllables and len(syllables[0]) > 2:
            complex_word_count += 1
    return complex_word_count

def total_word_count(url):
    response = requests.get(url)
    webpage_text = response.text
    words = word_tokenize(webpage_text)# Tokenize the text into words
    # Remove punctuation
    words = [word for word in words if word not in string.punctuation]
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.lower() not in stop_words]
    total_word_count = len(words)
    return total_word_count

def total_sentences_count(url):
    response = requests.get(url)
    webpage_text = response.text
    # Tokenize the text into sentences
    sentences = sent_tokenize(webpage_text)
    # Count the number of sentences
    sentence_count = len(sentences)
    return sentence_count

def sum_total_characters_per_word(url):
    response = requests.get(url)
    webpage_text = response.text
    # Tokenize the text into words
    words = word_tokenize(webpage_text)
    # Calculate the sum of total characters for all words
    total_character_count = sum(len(word) for word in words)
    return total_character_count

def calculate_positive_score(text):
    response = requests.get(url)
    webpage_text = response.text
    # Initialize the SentimentIntensityAnalyzer
    analyzer = SentimentIntensityAnalyzer()
    # Analyze sentiment for the text
    sentiment_scores = analyzer.polarity_scores(webpage_text)
    # Get the positive sentiment score (range: -1 to 1)
    positive_score = sentiment_scores['pos']
    return positive_score
def calculate_negative_score(text):
    response = requests.get(url)
    webpage_text = response.text
    # Initialize the SentimentIntensityAnalyzer
    analyzer = SentimentIntensityAnalyzer()
    # Analyze sentiment for the text
    sentiment_scores = analyzer.polarity_scores(webpage_text)
    # Get the positive sentiment score (range: -1 to 1)
    negative_score = sentiment_scores['neg']
    return negative_score


def count_syllables(word):
    word = word.lower()
    # Define exceptions where we won't count 'es' and 'ed' as syllables
    exceptions = ["es", "ed"]
    # Check for and remove trailing 'es' and 'ed' from the word
    for exception in exceptions:
        if word.endswith(exception):
            word = word[:-len(exception)]
    # Use the NLTK corpus words to check if the word is in the English vocabulary
    if word in nltk.corpus.words.words():
        # Count syllables based on vowels
        vowels = "aeiou"
        syllable_count = 0
        prev_char = None
        for char in word:
            if char in vowels and (prev_char is None or prev_char not in vowels):
                syllable_count += 1
            prev_char = char
        # Handle special cases where 'y' at the end of a word is considered a vowel
        if word.endswith("y") and prev_char not in vowels:
            syllable_count += 1
        # Ensure a minimum of one syllable for words with no detected syllables
        if syllable_count == 0:
            syllable_count = 1
    else:
        # If the word is not in the English vocabulary, return -1 to indicate an error
        syllable_count = -1
    return syllable_count

# Authorize with credentials
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
credentials = ServiceAccountCredentials.from_json_keyfile_name("data-scraping-399609-ea70628dccff.json", scope)
gc = gspread.authorize(credentials)

# Open a specific Google Sheet by its title or URL
worksheet = gc.open("Data_Crawling").sheet1

# Get all the values in the B column (assuming your URLs are in column B)
b_column_values = worksheet.col_values(2)[1:]  # The argument 2 represents column B (indexing is 1-based)//[1:]skips the 1st row(header)

# Print the URLs
for url in b_column_values:
    print("URL : ", url)
    avg_words=avg_words_per_sentence(url)
    print("Average Words per Sentence is :", avg_words)
    complex_word_count = count_complex_words(url)
    print("Complex words in url is :", complex_word_count)
    word_count=total_word_count(url)
    print("Word count in url is :", word_count)
    percentage_complex_words = (complex_word_count / word_count) * 100
    print("Percentage of complex words in url :", percentage_complex_words)
    sentence_count=total_sentences_count(url)
    print("Sentence count in url :",sentence_count)
    avg_sentence_length=(word_count/sentence_count)
    fog_index=0.4*(avg_sentence_length+percentage_complex_words)
    print("Average sentence length in url :",avg_sentence_length )
    print("Fog index in url : ", fog_index)
    #-------------------
    # response = requests.get(url)
    # webpage_text = response.text
    # words = word_tokenize(webpage_text)# Tokenize the text into words
    # total_syllable_count=0
    # for word in words:
    #   syllable_count = count_syllables(word)
    #   total_syllable_count += syllable_count
    # print('Syllable count in url:',total_syllable_count)
    #-------------------
    sum_char_per_word=sum_total_characters_per_word(url)
    avg_word_length=sum_char_per_word/word_count
    print("Average word length in url :", avg_word_length)
    positive_score=calculate_positive_score(url)
    print("Positive score in url :", positive_score)
    negative_score=calculate_positive_score(url)
    print("Negative score in url :", negative_score)
    polarity_score= (positive_score-negative_score)/((positive_score+negative_score)+0.000001)
    print("Polarity score in url :",polarity_score)
    subjectivity_score=(positive_score+negative_score)/((word_count)+0.000001)
    print("Subjectivity score in url :",polarity_score)








```
